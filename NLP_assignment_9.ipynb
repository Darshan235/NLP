## Step 3: Load Pre-trained Word2Vec Model
Using GoogleNews-vectors-negative300.bin.gz

```python
!wget -O /content/GoogleNews-vectors-negative300.bin.gz \
https://github.com/mmihaltz/word2vec-GoogleNews-vectors/raw/master/GoogleNews-vectors-negative300.bin.gz
```

```python
from gensim.models import KeyedVectors
import os

file_path = "/content/GoogleNews-vectors-negative300.bin.gz"

# Check if file exists
if not os.path.exists(file_path):
    raise FileNotFoundError("GoogleNews Word2Vec file not found in /content")

# Load the model
model = KeyedVectors.load_word2vec_format(
    file_path,
    binary=True
)

print("Model loaded successfully")
print("Vocabulary size:", len(model.key_to_index))
```

## Example Word Vector

```python
model['king'][:10]
```

## Step 4: Word Similarity

```python
pairs = [
    ("doctor", "nurse"),
    ("cat", "dog"),
    ("car", "bus"),
    ("king", "queen"),
    ("apple", "banana"),
    ("man", "woman"),
    ("teacher", "student"),
    ("computer", "keyboard"),
    ("sun", "moon"),
    ("india", "china")
]

for w1, w2 in pairs:
    print(f"{w1} - {w2}: {model.similarity(w1, w2):.3f}")
```

## Step 5: Nearest Neighbors

```python
words = ["king", "university", "doctor", "computer", "india"]

for w in words:
    print(f"\nNearest words to '{w}':")
    for word, score in model.most_similar(w, topn=5):
        print(word, score)
```

## Step 6: Word Analogies

```python
print("king - man + woman =", model.most_similar(positive=["king","woman"], negative=["man"], topn=1))
print("paris - france + india =", model.most_similar(positive=["paris","india"], negative=["france"], topn=1))
print("teacher - school + hospital =", model.most_similar(positive=["teacher","hospital"], negative=["school"], topn=1))
```
